```python
import torch
from transformers import MarianMTModel, MarianTokenizer

device = torch.device('cuda' if
                      torch.cuda.is_available() else 'cpu')

print(device)

def translate_marianMt(text, model_name="Helsinki-NLP/opus-mt-en-id"):
    # Load pre-trained model and tokenizer
    model = MarianMTModel.from_pretrained(model_name).to(device)
    tokenizer = MarianTokenizer.from_pretrained(model_name)

    # Tokenize input text
    input_ids = tokenizer.encode(text, return_tensors="pt").to(device)

    # Generate translation
    translation_ids = model.generate(input_ids)

    # Decode the translation
    translation = tokenizer.decode(translation_ids[0], skip_special_tokens=True)

    return translation

# Example usage
english_text = "Hello, how are you?"
indonesian_translation = translate_marianMt(english_text)

print(f"English: {english_text}")
print(f"Indonesian Translation: {indonesian_translation}")
```

```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

def translate_MT5(text_en, model_name="acul3/mt5-translate-en-id"):
  tokenizer = AutoTokenizer.from_pretrained(model_name, use_fast=False)
  model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

  input_ids = tokenizer(text_en, return_tensors="pt").input_ids

  outputs = model.generate(input_ids)
  translated_text = tokenizer.batch_decode(outputs[0], skip_special_tokens=True, clean_up_tokenization_spaces=False)

  return translated_text
```

```python
from googletrans import Translator
#from googleapiclient.discovery import build

def translate_google_text(text):
  try:
    translator = Translator()
    translated = translator.translate(text, src='en', dest='id').text
    return translated
  except Exception as e:
    print(f"Error: {e}")
    return ''

```

```python
import dl_translate as dlt

def translate_m2m100(text_en):
  mt = dlt.TranslationModel("facebook/m2m100_418M")
  translation_id = mt.translate(text_en, source=dlt.lang.ENGLISH, target=dlt.lang.INDONESIAN)
  return translation_id

def translate_mbart50(text_en):
  mt = dlt.TranslationModel("facebook/mbart-large-50-many-to-many-mmt")
  translation_id = mt.translate(text_en, source=dlt.lang.ENGLISH, target=dlt.lang.INDONESIAN)
  return translation_id

def translate_nllb200(text_en):
  mt = dlt.TranslationModel("facebook/nllb-200-distilled-600M")
  translation_id = mt.translate(text_en, source=dlt.lang.ENGLISH, target=dlt.lang.INDONESIAN)
  return translation_id
```

```python
import pandas as pd
import nltk
from nltk.translate.bleu_score import sentence_bleu

nltk.download('punkt')
# Assuming df is your DataFrame and it has been loaded correctly
# Make sure to preprocess your text data as necessary
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/IDENTICv1.0/val_google.csv')


# Function to calculate BLEU score for each row
def calculate_bleu_for_row(row):
    reference = nltk.word_tokenize(row['id_text'])
    candidate = nltk.word_tokenize(row['new_id_text'])
    return sentence_bleu([reference], candidate)

# Apply the function to each row and create a new column 'bleu_score'
df['bleu_score'] = df.apply(calculate_bleu_for_row, axis=1)

# Display the DataFrame with the BLEU scores
print(df[['id_text', 'new_id_text', 'bleu_score']])
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import torch

def load_fine_tuned_model(model_path):
    """
    Load the fine-tuned model and tokenizer from the specified path.

    Args:
        model_path (str): The directory path where the model and tokenizer are saved.

    Returns:
        model: The fine-tuned model.
        tokenizer: The tokenizer corresponding to the model.
        device: The device (CPU or GPU) on which the model will be run.
    """
    tokenizer = AutoTokenizer.from_pretrained(model_path)
    model = AutoModelForSeq2SeqLM.from_pretrained(model_path)
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)
    return model, tokenizer, device

def translate_finetune_nllb200(model, tokenizer, device, input_text, max_length=128):
    """
    Generate text using the fine-tuned model and tokenizer.

    Args:
        model: The fine-tuned model.
        tokenizer: The tokenizer corresponding to the model.
        device: The device (CPU or GPU) on which the model is run.
        input_text (str): The input text to be processed by the model.
        max_length (int): The maximum length of the generated text. Default is 128.

    Returns:
        str: The generated output text.
    """
    # Tokenize the input text
    inputs = tokenizer(input_text, return_tensors="pt", padding=True, truncation=True, max_length=max_length)
    inputs = {key: value.to(device) for key, value in inputs.items()}

    # Generate the output sequence
    with torch.no_grad():
        generated_ids = model.generate(**inputs, max_length=max_length)

    # Decode the generated tokens
    output_text = tokenizer.decode(generated_ids[0], skip_special_tokens=True)
    return output_text

# Define the path to your fine-tuned model directory
model_path = "/content/drive/MyDrive/Colab Notebooks/FineTune/fine-tuned-nllb-200"

# Load the model and tokenizer
model, tokenizer, device = load_fine_tuned_model(model_path)

# Example input text
input_text = "A man with a red helmet on a small moped on a dirt road."

# Generate output text
output_text = translate_finetune_nllb200(model, tokenizer, device, input_text)

print("Generated Output:", output_text)
```
